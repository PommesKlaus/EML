```python
def read_data(D):
    X = []
    y = []
    for elem in D:
        X.append(list(elem[0]))
        y.append(elem[1])
    return np.array(X), np.array(y)

def read_2d_df(D):
    data = [[t[0][0], t[0][1], t[1]] for t in D]
    df = pd.DataFrame(data)
    return df.iloc[:,0:-1].values, df.iloc[:,-1].values, df
```

## Frage 1

```python
from sklearn.tree import DecisionTreeClassifier

D_train={((37.5,2),1),((36,1),0),((39,1),1),((37.5,1),0),((39,0),1),((37.5,0),0),((37.5,1),0),((36,2),1)}
D_test={((36,0),0),((37.5,1),0),((39,1),1),((39,2),0)}
X_train, y_train, df_train = read_2d_df(D_train)
X_test, y_test, df_test = read_2d_df(D_test)

tree = DecisionTreeClassifier().fit(X_train, y_train)
y_pred = tree.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics
print(y_pred)
print(y_test)
print(metrics.confusion_matrix(y_test, y_pred))

cm = metrics.confusion_matrix(y_test, y_pred)
confusion_matrix = [[cm[1][1], cm[0][1]], [cm[1][0], cm[0][0]]]
print(np.array(confusion_matrix))
```

## Frage 2

```python
from sklearn.ensemble import RandomForestClassifier

D_train={((37.5,2),1),((36,1),0),((39,1),1),((37.5,1),0),((39,0),1),((37.5,0),0),((37.5,1),0),((36,2),1)}
D_test={((36,0),0),((37.5,1),0),((39,1),1),((39,2),0)}
X_train, y_train, df_train = read_2d_df(D_train)
X_test, y_test, df_test = read_2d_df(D_test)

forest = RandomForestClassifier(n_estimators=100)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

print("Genauigkeit:", metrics.accuracy_score(y_test, y_pred))
print("F1:", metrics.f1_score(y_test, y_pred))
```

```python
for t in [10, 1000]:
    print("t=",t)
    forest = RandomForestClassifier(n_estimators=t)
    forest.fit(X_train, y_train)
    y_pred = forest.predict(X_test)
    print(y_pred)
    print("Genauigkeit:", metrics.accuracy_score(y_test, y_pred))
    print("F1:", metrics.f1_score(y_test, y_pred))
    print()
```
