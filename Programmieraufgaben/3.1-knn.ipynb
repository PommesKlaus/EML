```python
from sklearn.cluster import KMeans

E = [(8.5,16),(12.4,26.8),(9.3,21),(11.5,17),(17.5,25),(13,22.6),(4,16.5)]
X = np.array(E)

knn2 = KMeans(n_clusters=2)
knn2.fit_predict(X)
```

```python
knn2.cluster_centers_
```

```python
knn2.inertia_
```

```python
knn3 = KMeans(n_clusters=3)
knn3.fit_predict(X)
```

```python
np.round(knn3.cluster_centers_,3)
```

```python
knn3.inertia_
```
