```python
def read_data(D):
    X = []
    y = []
    for elem in D:
        X.append(list(elem[0]))
        y.append(elem[1])
    return np.array(X), np.array(y)
```

## Frage 1

```python
from sklearn.naive_bayes import GaussianNB

D={((0,50),0),((1,200),2),((2,200),1),((0,100),1),((1,50),0),((0,200),2)}
X_train, y_train = read_data(D)

gnb = GaussianNB()
gnb.fit(X_train, y_train)

res = gnb.predict(np.array([2,50]).reshape(1, -1))
```

## Frage 2

```python
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

D_train = {((1.29,-1.83),1),((0.76,-0.95),1),((5.24,-1.59),0),((8.74,-0.27),0),((2.77,-2.23),0),((1.24,-1.65),1)}
X_train, y_train = read_data(D_train)

D_test={((0.62,-1.58),1),((6.82,0.07),0),((2.64,-0.15),1),((3.23,-1.58),0)}
X_test, y_test = read_data(D_test)

gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)

for i, t in enumerate(X_test):
    print(t, y_pred[i])
print("Genauigkeit=", metrics.accuracy_score(y_test, y_pred))
print("Pr채zision=", metrics.precision_score(y_test, y_pred))
print("Sensitivit채t=", metrics.recall_score(y_test, y_pred))
print("F1=", metrics.f1_score(y_test, y_pred))
```
