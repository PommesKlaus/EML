```python
E=[(11,0.8),(11.6,1),(9.2,0.9),(14.1,0.8),(13.4,0.9),(8.8,1.1),(12.2,0.8)]
X = np.array(E)
X
```

```python
from sklearn.covariance import EllipticEnvelope
ell = EllipticEnvelope()
ell.fit(X)
ell.predict(np.array([[13.9,1.1]]))
```

```python
from matplotlib import pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay

disp = DecisionBoundaryDisplay.from_estimator(ell, X, response_method='predict')

plt.scatter(X[:,0], X[:,1], marker="x")
plt.scatter(13.9, 1.1, marker="o")
```
